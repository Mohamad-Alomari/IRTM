```python
def remove_stop_words(text):
    filtered_sentence= []
    for w in text: 
        
        if w not in stop_words: 
            filtered_sentence.append(w) 
            
    return filtered_sentence

def stemming(text):
    stemmed_sentence= []
    for w in text: 
        w_stemming = ps.stem(w)
        stemmed_sentence.append(w_stemming) 
    return stemmed_sentence
```
