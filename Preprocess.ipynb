```python
import re
from nltk.tokenize import RegexpTokenizer
```

```python
data = open("script/10_01.txt", "r") 
text_first = data.read()
```

```python
text = "\n".join([ll.rstrip() for ll in text_first.splitlines() if ll.strip()])
print(text)
```

```python
main_char = ["Leonard","Sheldon","Penny", "Howard","Raj","Amy","Bernadette"]
lines= {'Leonard': [], 'Sheldon': [], 'Penny': [], 'Howard': [], 'Raj': [],
                   'Amy': [], 'Bernadette': []}
```

```python
tt = text.split("\n")
for t in tt[1:]:
    try:
        print(t)
        (speaker, dialogue) =  t.split(":")
        speaker = speaker.strip()
        speaker = speaker.split("(")[0].strip()
        tokenizer = RegexpTokenizer(r'\w+')
        words = tokenizer.tokenize(dialogue.lower())
        if speaker in main_char:
            lines[speaker].append(words)
    except ValueError:
        pass
        
        
        
    
```
